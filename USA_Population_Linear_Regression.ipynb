```python
import pandas as pd
import matplotlib.pyplot as mplib
from sklearn.linear_model import LinearRegression 
```

```python
Data = pd.read_csv('USA_Population_1960-2020.csv')
```

```python
Year = pd.DataFrame(Data,columns=["Year"])
Population = pd.DataFrame(Data,columns=["Population"])
```

```python
mplib.figure(figsize=(20.5,8.5))
mplib.scatter(Year,Population,alpha=0.6)
mplib.title("USA Population 1960-2020")
mplib.xlabel("Year")
mplib.ylabel("Population")
mplib.ylim(15000000,350000000)
mplib.xlim(1959,2021)
mplib.style.use("fivethirtyeight")
```

```python
LinearRegression  = LinearRegression()
LinearRegression.fit(Year,Population)
LinearRegression.coef_ # returns slope coeficent theta_subcript_1
```

```python
LinearRegression.intercept_
```

```python
LinearRegression.score(Year,Population)
```
